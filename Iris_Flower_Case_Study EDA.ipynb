```python
df3 = df.drop(columns=['SepalRation',	'PetalRation','Id'],axis=1)
df3
```

```python
df3['Species'].unique()
```

```python
df3['Species'].replace('Iris-setosa',0,inplace=True)
df3['Species'].replace('Iris-versicolor',1,inplace=True)
df3['Species'].replace('Iris-virginica',2,inplace=True)
```

```python
df3.head()
```

```python
#Step-1) Choose Independent & Dependent Variables
x = df3[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y = df3[['Species']]
```

```python
print(x)
```

```python
print(y)
```

```python
#Step-2) Split dataset into training & testing shape
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x,y,train_size=0.75)
```

```python
#Step-3) Shape
print(df3.shape)
print(x_train.shape,x_test.shape)
print(y_train.shape,y_test.shape)
```

```python
#Step-4) ML algorithm build up - Classification - LogisticRegression
from sklearn.linear_model import LogisticRegression 
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
#Step-5) Score Check - Testing
model.score(x_test,y_test)
```

```python
#Step-6) Finding Predicted Values
y_pred = model.predict(x_test)
y_pred[0:5]
```

```python
#Step-7) Actual Values
y_test.head()
```

```python
#Real Prediction 
newflower = np.array([5.1  ,          3.5    ,         1.4    ,        0.2]).reshape(1,-1)
newflower
```

```python
r_pred = model.predict(newflower)
r_pred
```

```python
if(r_pred[0]==0):
    print("Setosa")
elif(r_pred[0]==1):
    print("versicolor")
else:
    print("virginica")
```
